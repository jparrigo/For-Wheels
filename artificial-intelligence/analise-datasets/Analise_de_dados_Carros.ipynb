```python
import pandas as pd
import numpy as np

df_car = pd.read_csv("../datasets/cars/data.csv", sep=",", header=0)
df_car.head()
```

```python
df_car.info()
```

```python
brands = ["Saab", "Scion"]
minYear = 2010
gasoline_fuel = [
    "gasoline",
    "regular unleaded",
    "premium unleaded (required)",
    "premium unleaded (recommended)",
]
flex_fuel = [
    "flex-fuel",
    "flex-fuel (unleaded/E85)",
    "flex-fuel (premium unleaded required/E85)",
    "flex-fuel (premium unleaded recommended/E85)",
]
natGas = ["natural gas", "flex-fuel (unleaded/natural gas)"]
suv = ["SUV", "4dr SUV", "2dr SUV", "Convertible SUV"]
hatchback = ["Hatchback", "4dr Hatchback", "2dr Hatchback"]
pickup = ["Pickup", "Crew Cab Pickup", "Extended Cab Pickup", "Regular Cab Pickup"]
minivan = ["Minivan", "Passenger Minivan", "Cargo Minivan"]
fwd = ["front wheel drive", "FWD"]
rwd = ["rear wheel drive", "RWD"]
awd = ["all wheel drive", "AWD"]
fourwd = ["four wheel drive", "4WD"]
three_dr = 3
df_car = df_car.where(df_car["Make"] != brands[0])
df_car = df_car.where(df_car["Make"] != brands[1])
df_car = df_car.where(df_car["Year"] >= minYear)
df_car = df_car.where(df_car["Engine Fuel Type"] != natGas[0])
df_car = df_car.where(df_car["Engine Fuel Type"] != natGas[1])
df_car = df_car.where(df_car["Number of Doors"] != three_dr)
df_car["Engine Fuel Type"] = np.where(
    df_car["Engine Fuel Type"] == gasoline_fuel[1],
    gasoline_fuel[0],
    df_car["Engine Fuel Type"],
)
df_car["Engine Fuel Type"] = np.where(
    df_car["Engine Fuel Type"] == gasoline_fuel[2],
    gasoline_fuel[0],
    df_car["Engine Fuel Type"],
)
df_car["Engine Fuel Type"] = np.where(
    df_car["Engine Fuel Type"] == gasoline_fuel[3],
    gasoline_fuel[0],
    df_car["Engine Fuel Type"],
)
df_car["Engine Fuel Type"] = np.where(
    df_car["Engine Fuel Type"] == flex_fuel[1], flex_fuel[0], df_car["Engine Fuel Type"]
)
df_car["Engine Fuel Type"] = np.where(
    df_car["Engine Fuel Type"] == flex_fuel[2], flex_fuel[0], df_car["Engine Fuel Type"]
)
df_car["Engine Fuel Type"] = np.where(
    df_car["Engine Fuel Type"] == flex_fuel[3], flex_fuel[0], df_car["Engine Fuel Type"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == suv[1], suv[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == suv[2], suv[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == suv[3], suv[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == hatchback[1], hatchback[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == hatchback[2], hatchback[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == pickup[1], pickup[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == pickup[2], pickup[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == pickup[3], pickup[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == minivan[1], minivan[0], df_car["Vehicle Style"]
)
df_car["Vehicle Style"] = np.where(
    df_car["Vehicle Style"] == minivan[2], minivan[0], df_car["Vehicle Style"]
)
df_car["Driven_Wheels"] = np.where(
    df_car["Driven_Wheels"] == fwd[0], fwd[1], df_car["Driven_Wheels"]
)
df_car["Driven_Wheels"] = np.where(
    df_car["Driven_Wheels"] == rwd[0], rwd[1], df_car["Driven_Wheels"]
)
df_car["Driven_Wheels"] = np.where(
    df_car["Driven_Wheels"] == awd[0], awd[1], df_car["Driven_Wheels"]
)
df_car["Driven_Wheels"] = np.where(
    df_car["Driven_Wheels"] == fourwd[0], fourwd[1], df_car["Driven_Wheels"]
)
df_car = df_car.dropna(axis=0, how="all")
df_car = df_car.reset_index()
df_car.drop("index", axis=1, inplace=True)
df_car.head()
```

```python
df_car.info()
```

```python
df_car["Market Category"] = df_car["Market Category"].fillna("Common")
df_car.head()
```

```python
temp = df_car
temp[["Category 1", "Category 2", "Category 3", "Category 4", "Category 5"]] = df_car[
    "Market Category"
].str.split(",", expand=True)
temp.head()
```

```python
temp_dummy1 = pd.get_dummies(temp["Category 1"])
temp_dummy2 = pd.get_dummies(temp["Category 2"])
temp_dummy3 = pd.get_dummies(temp["Category 3"])
temp_dummy4 = pd.get_dummies(temp["Category 4"])
temp_dummy5 = pd.get_dummies(temp["Category 5"])
```

```python
temp_dummy1 = temp_dummy1.reindex(sorted(temp_dummy1.columns), axis=1)
temp_dummy2 = temp_dummy2.reindex(sorted(temp_dummy1.columns), axis=1)
temp_dummy3 = temp_dummy3.reindex(sorted(temp_dummy1.columns), axis=1)
temp_dummy4 = temp_dummy4.reindex(sorted(temp_dummy1.columns), axis=1)
temp_dummy5 = temp_dummy5.reindex(sorted(temp_dummy1.columns), axis=1)
```

```python
category_dummy = temp_dummy1 | temp_dummy2 | temp_dummy3 | temp_dummy4 | temp_dummy5
category_dummy.info()
```

```python
old_df_car = df_car
df_car = pd.concat([old_df_car.drop("Market Category", axis=1), category_dummy], axis=1)
df_car = df_car.drop(
    [
        "Category 1",
        "Category 2",
        "Category 3",
        "Category 4",
        "Category 5",
        "Diesel",
        "Flex Fuel",
    ],
    axis=1,
)
display(df_car)
```

```python
city_mean = np.mean(df_car["city mpg"])
df_car["City Preference"] = df_car["city mpg"] > city_mean
df_car["Highway Preference"] = df_car["City Preference"] != True
```

```python
df_car = df_car.astype(
    {
        "Year": "int64",
        "Engine HP": "int64",
        "Engine Cylinders": "int64",
        "Number of Doors": "int64",
        "highway MPG": "int64",
        "city mpg": "int64",
        "Popularity": "int64",
        "MSRP": "int64",
    }
)
df_car["Vehicle_Title_Partial"] = (
    df_car["Year"].astype(str)
    + " "
    + df_car["Make"].astype(str)
    + " "
    + df_car["Model"].astype(str)
    + " "
    + df_car["Vehicle Style"]
)
df_car["delete"] = (
    df_car["Year"].astype(str)
    + " "
    + df_car["Make"].astype(str)
    + " "
    + df_car["Model"].astype(str)
    + " "
    + df_car["Vehicle Style"]
    + " "
    + df_car["Transmission Type"]
    + " "
    + df_car["Engine Fuel Type"]
    + " "
    + df_car["Vehicle Size"]
)
df_car["delete2"] = (
    df_car["Year"].astype(str)
    + " "
    + df_car["Make"].astype(str)
    + " "
    + df_car["Model"].astype(str)
    + " "
    + df_car["Vehicle Style"]
    + " "
    + df_car["Engine HP"].astype(str)
)
```

```python
df_car["delete"].nunique()
```

```python
df_car.drop_duplicates(subset="delete", keep="first", inplace=True)
df_car.drop_duplicates(subset="delete2", keep="first", inplace=True)
df_car.drop("delete", axis=1, inplace=True)
df_car.drop("delete2", axis=1, inplace=True)
df_car = df_car.dropna(axis=0, how="all")
df_car = df_car.reset_index()
df_car.drop("index", axis=1, inplace=True)
```
